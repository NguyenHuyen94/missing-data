```python
import pandas as pd
import numpy as np
df_train=pd.read_csv("train.csv")
df_train.head()
```

```python
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
#nhận xét tổng quan structure dât
# ta có 2 set data "train", "set" ở dạng giống nhau (các cột hàng, cấu trúc giống nhau)
#tuy nhiên "train" có thêm cột survied -> CẦN FIX
#nối lại thành 1 bảng để phân tích
# dùng lệnh pandas.concat
a=len(df_train)
a
```

```python

#drop luôn cái cột survied để cho cùng dạng với "set"
dftrain = df_train.drop('Survived', axis=1)
dftrain
```

```python
#giờ ghép 2 data với nhau lệnh concat
df= pd.concat([df_test, dftrain],axis = 0)
df
```

```python
#kiểm tra lại dữ liệu 1 lần nữa 
df.info()
```

```python
#kiểm tra nhưng cột có data bị thiếu 
df.isnull().sum().sort_values(ascending=False)
```

```python
#2 cách xử lý:
#c1: drop nó đi (nếu nó bị thiếu ít dưới 5% hoặc nhiều nhưng không có giá trị trong phân tích)
#c2: fill bằng mean(giá trị chia đồ thị thành 2 phần đều nhau), median (giá trị trung bình)
```

```python
#đối với data này, ta drop fare null, embardked null
df = df.dropna(subset = ['Fare', 'Embarked'])
df.isnull().sum()
```

```python
#có thể sử dụng thư viện missingno để tìm ra tỷ lệ null là bao nhiêu %, nó sẽ visualize giá trị null
import missingno as msno
import matplotlib.pyplot as plt
plt.figure(figsize = (1,2))
data_missings = df.filter(['Age','Cabin'], axis = 1)
msno.bar(data_missings)
```

```python
#thay thế các giá trị bị mất này 
# cách 1: 
#Với những features có tính liên tục thì việc chúng ta thay thế những giá trị missing bằng các giá trị -1, -99, -999, ..
# Nhược điểm của nó làm giải hiệu suất của mô hình tuyến tính sẽ bị ảnh hưởng.
#cách 2:
#Thay thế bằng giá trị mean, midian, mode
#nếu biến liên tục: mean (phổ biến hơn), median
# nếu biến là catergory: mode
# phù hợp với mô hình tuyến tính đơn giản và NN. 
# Nhược điểm: làm giảm phương sai của dữ liệu.
# những bài toán dựa trên tree thì có vẻ không phù hợp lắm.
```

```python
df.Age.fillna(df.Age.mean(),inplace=True)
df.Cabin.fillna(df.Cabin.mode()[0],inplace=True)
df.isnull().sum()
```
